* Now, rather than having to call each of these functions separately to get the three measures of central tendency, write a custom function to return all three values in a single function call:

1. Name the function "central_tendency" and start with a single positional argument that accepts the list of temperatures.
2. Define three variables as mean_temp, median_temp, and mode_temp using the existing functions from NumPy and the statistics module.
3. Add a return statement to return all three variables.

```python
# Write a custom function to return all 3 values in a single call
def central_tendency(week):
  mean_temp = np.mean(week)
  median_temp = np.median(week)
  mode_temp = mode(week)
  return mean_temp, median_temp, mode_temp
```

If verbose = False, the function will just return the mean, median, and mode values without the print statements.

```python
# Create a modified version to include keyword argument for verbose
def central_tendency(week, verbose = True):
  mean_temp = np.mean(week)
  median_temp = np.median(week)
  mode_temp = mode(week)
  if verbose == True:
    print(f'The mean is {mean_temp}. The median is {median_temp}. The mode is {mode_temp}.')
  else:
    return mean_temp, median_temp, mode_temp
```
